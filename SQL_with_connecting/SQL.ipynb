## Составление запросов для поставленных задач

### Cколько книг вышло после 1 января 2000 года;

```python
query_1 = ''' 
    SELECT COUNT(*) 
    FROM books
    WHERE publication_date > '2000-01-01'; 
''' 
pd.io.sql.read_sql(sql=text(query_1), con = con)
```

**Располагаем данными о 819 книгах, выпущенных в текущем тысячелетии**

### Для каждой книги посчитайте количество обзоров и среднюю оценку

```python
query_2 = '''WITH rev_cnt AS 
                (SELECT book_id,
                        COUNT (review_id) AS review_cnt
                FROM reviews
                GROUP BY 1),   
    
            avg_rat AS 
                (SELECT book_id,
                        AVG (rating) AS rating_avg
                FROM ratings
                GROUP BY 1) 
       
                SELECT b.title,
                        rev_cnt.review_cnt,
                        avg_rat.rating_avg
                FROM books AS b 
                LEFT JOIN rev_cnt ON b.book_id=rev_cnt.book_id  
                LEFT JOIN avg_rat ON b.book_id=avg_rat.book_id
'''

pd.io.sql.read_sql(sql=text(query_2), con = con)
```

**Видим пропуски в review_cnt**. Проверим корректность выполнения первого подзапроса в WITH: для всех ли книг есть рецензии?

```python
review_cnt = '''SELECT book_id,
        COUNT (review_id) AS review_cnt
        FROM reviews
    GROUP BY 1'''
pd.io.sql.read_sql(sql=text(review_cnt), con = con)['book_id'].nunique()
```

На 6 книг нет отзывов, запрос работает корректно.

### Определите издательство, которое выпустило наибольшее число книг толще 50 страниц

```python
query_3 = ''' WITH top AS
                (SELECT publisher_id,
                        COUNT (book_id) AS book_cnt
                FROM books
                WHERE num_pages > 50
                GROUP BY 1
                ORDER BY 2 DESC
                LIMIT 1)
                
                SELECT p.publisher
                FROM publishers AS p
                RIGHT JOIN top ON p.publisher_id = top.publisher_id
'''
pd.io.sql.read_sql(sql=text(query_3), con = con)
```

**Больше всего книг толще 50 страниц выпустило издательство "Penguin Books"**

### Определите автора с самой высокой средней оценкой книг — учитывайте только книги с 50 и более оценками

```python
query_4 = '''WITH top_cnt AS
                (SELECT book_id,
                    COUNT(rating) AS rating_cnt,
                    AVG(rating) AS avg_rating
                FROM ratings
                GROUP BY 1
                HAVING  COUNT(rating) >= 50)
    
            SELECT author,
                AVG(avg_rating) AS avg_rating
            FROM books AS b
            RIGHT JOIN top_cnt ON top_cnt.book_id= b.book_id
            LEFT JOIN authors AS a ON b.author_id = a.author_id
            GROUP BY author
            ORDER BY 2 DESC
            LIMIT 1
'''
pd.io.sql.read_sql(sql=text(query_4), con = con)
```

**Самая высокая средняя оценка книг у J.K. Rowling/Mary GrandPré**

### Посчитайте среднее количество обзоров от пользователей, которые поставили больше 48 оценок

```python
query_5 = '''SELECT AVG(rev_cnt)
            FROM(
                SELECT username,
                        COUNT(review_id) AS rev_cnt
                FROM reviews 
                WHERE username IN (SELECT username
                                    FROM ratings
                                    GROUP BY 1
                                    HAVING  COUNT(rating_id) > 48)
                                    GROUP BY 1)
                                    AS avg_rev_cnt
    
'''
pd.io.sql.read_sql(sql=text(query_5), con = con)
```

**Среди активных пользователей, поставивших более 48 оценок, среднее количество обзоров: 48**
